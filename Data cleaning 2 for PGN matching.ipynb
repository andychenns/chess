```python
import pandas as pd
import numpy as np
```

```python
#Load CSV file
df = pd.read_csv (r'/Users/longhaochen/chess/chess/data/lichess-04-2017_movebreakdown.csv')
```

```python
df.head(5)
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.iloc[1244,]
```

```python
df = df.assign(Move=df.Move.str.split(),Eval=df.Eval.str.split(),Clock=df.Clock.str.split())
```

```python
df = df.explode(["Move","Eval","Clock"])
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.head(100)
```

```python
df['Move'] = df['Move'].map(lambda x: x.lstrip("['").rstrip("]'"))
```

```python
df['Eval'] = df['Eval'].map(lambda x: x.lstrip("['").rstrip("]'"))
```

```python
df['Clock'] = df['Clock'].map(lambda x: x.lstrip("['").rstrip("]'"))
```

```python
df.to_csv("~/chess/chess/data/lichess-04-2017_movebreakdown2.csv",  header=True, index=False)
```
