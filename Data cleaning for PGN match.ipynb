```python
#First define a function that will decompoese a single PGN game then use list comprehension to loop through each PGN

PGN_output = pd.DataFrame(columns = ["ID", "Move", "Eval", "Clock"])
def PGN_Decompose(PGN, ID):
    #Load PGN data
    text_break = pd.Series(PGN.split(' '))
    #text_break = text_break.iloc[0].squeeze()
    #Search for data that we need
    eval_index = text_break[text_break.str.find("\[%eval")].index
    #move_number = text_break[eval_index-3]
    move = text_break[eval_index-2]
    evaluation = text_break[eval_index+1]
    clock_time = text_break[eval_index+3]
    #Reset index for all columns
    move.reset_index(drop=True, inplace=True)
    #move_number.reset_index(drop=True, inplace=True)
    evaluation.reset_index(drop=True, inplace=True)
    clock_time.reset_index(drop=True, inplace=True)
    #Generate series for ID for each move that we have
    match_id = pd.Series([ID] *len(move))
    #Concatenate columns
    temp = pd.concat([match_id,move,evaluation,clock_time],axis=1,ignore_index=True)
    temp.columns = ["ID", "Move", "Eval", "Clock"]
    #Concatenate temp to global dataframe
    #PGN_output = PGN_output.append(temp)
    return temp
```
